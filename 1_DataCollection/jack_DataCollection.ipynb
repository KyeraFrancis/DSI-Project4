# Scraping for player data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# URL to scrape
url = "https://www.footballguys.com/stats/season/teams?team=ARI&year=2023"

# Send an HTTP request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
# Find the table containing the player data
table = soup.find('table')

# Extract headers (column names) from the table
headers = [th.text.strip() for th in table.find_all('th')]

# Initialize an empty list to store rows
rows = []

# Extract data from each row in the table
for tr in table.find_all('tr')[0:]:  # Skip the header row
    row_data = [td.text.strip() for td in tr.find_all('td')]
    position = row_data[2]  # Assuming position is in the third column, adjust accordingly
    row_data.append(position)  # Add a 'Position' column
    rows.append(row_data)
```

```python
# Create a pandas DataFrame
df = pd.DataFrame(rows, columns=headers + ['Position'])

# Display the DataFrame
print(df.head())
```
