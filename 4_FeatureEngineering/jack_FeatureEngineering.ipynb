### Transforming multiple columns

```python
import pandas as pd

# Read in the CSV 
df = pd.read_csv('../2_DataCleaning/cleaned_data/clean_merged_data.csv')

# Drop unnecessary columns
df = df.drop(columns=['Unnamed: 0'])

# Groupby player ID
df = df.groupby('gsis_id', as_index=False).last()

# Get count of each report status per group
status_counts = df.groupby('gsis_id')['report_status_x'].value_counts().unstack()

# Rename columns 
status_counts.columns = [f'num_{col}' for col in status_counts.columns]

# Join the count columns back to the df
df = df.join(status_counts)

df
```
